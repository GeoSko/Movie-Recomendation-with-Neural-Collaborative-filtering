```python
import pandas as pd
import numpy as np
import matplotlib as plt

df = pd.read_csv('u.data', sep = '\t') 
df.columns = ['user','movie','rate','timestamp']              # ονομάζω στήλες
df = df.drop(['timestamp'], axis = 1)                         # αφαιρώ το timestamp
df = df.pivot(index='user', columns='movie', values='rate')   # φέρνω το αρχείο εγγραφών σε μορφή πίνακα
df.head()
```

```python
means = df.mean(axis=1)
means.head()
```

```python
%matplotlib inline
mean_plot = means.plot.hist(bins=20 )
mean_plot.set_xlabel("User's rating mean")
mean_plot.set_ylabel("Number of users")
```

```python
stds = df.std(axis=1)
mean_plot = stds.plot.hist(bins=20 )
mean_plot.set_xlabel("User's rating std")
mean_plot.set_ylabel("Number of users")
```

```python
df= df.sub(df.mean(axis=1),axis=0)
df.head()
```

```python
df.fillna(value=0, inplace= True)
df.head()
```

```python
df = df.add(abs(df.min().min()))
df.head()
```

```python
users_one_hot = pd.get_dummies(df.index)
users_one_hot.head()
```

```python
df.to_csv('output.csv', sep=',', index=False, header= False)
users_one_hot.to_csv('input.csv', sep=',', index=False, header= False)
print("Created csv files")
```
